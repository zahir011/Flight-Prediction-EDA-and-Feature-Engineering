```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Create an instance of OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)

# Fit and transform the 'Airline' column
airline_encoded = ohe.fit_transform(final_df[['Airline']])
final_df = final_df.join(pd.DataFrame(airline_encoded, columns=ohe.get_feature_names_out(['Airline'])))

# Fit and transform the 'Source' column
source_encoded = ohe.fit_transform(final_df[['Source']])
final_df = final_df.join(pd.DataFrame(source_encoded, columns=ohe.get_feature_names_out(['Source'])))

# Fit and transform the 'Destination' column
destination_encoded = ohe.fit_transform(final_df[['Destination']])
final_df = final_df.join(pd.DataFrame(destination_encoded, columns=ohe.get_feature_names_out(['Destination'])))

# Fit and transform the 'Additional_Info' column
additional_info_encoded = ohe.fit_transform(final_df[['Additional_Info']])
final_df = final_df.join(pd.DataFrame(additional_info_encoded, columns=ohe.get_feature_names_out(['Additional_Info'])))

# Drop the original columns if needed
final_df.drop(['Airline', 'Source', 'Destination', 'Additional_Info'], axis=1, inplace=True)

final_df.head()
```
